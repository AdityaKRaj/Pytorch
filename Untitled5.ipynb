```python
import pandas as pd
import numpy as np
import random
import tensorflow as tf
import matplotlib.pyplot as plt
from matplotlib import cm
import os








# Load CVRPTW Dataset
data_dir = '/content'  # Ensure this path is correct
orders_file = os.path.join(data_dir, 'orders.csv')
orders_df = pd.read_csv(orders_file)
orders_df = orders_df.head(10)  # Limit to 25 for now








# Preprocess the dataset (normalize coordinates)
def normalize_data(orders_df, grid_max=100):
    min_x, max_x = orders_df['x'].min(), orders_df['x'].max()
    min_y, max_y = orders_df['y'].min(), orders_df['y'].max()
    max_range = max(max_x - min_x, max_y - min_y)
    orders_df['x'] = ((orders_df['x'] - min_x) / max_range) * grid_max
    orders_df['y'] = ((orders_df['y'] - min_y) / max_range) * grid_max
    return orders_df








orders_df = normalize_data(orders_df)




# Split the dataset into 75% training and 25% testing
train_size = int(0.75 * len(orders_df))
train_df = orders_df[:train_size]
test_df = orders_df[train_size:]












# QuadTree class for efficient spatial querying
class QuadTree:
    def __init__(self, boundary, capacity):
        self.boundary = boundary  # (x_min, x_max, y_min, y_max)
        self.capacity = capacity
        self.agents = []
        self.divided = False




    def subdivide(self):
        x_min, x_max, y_min, y_max = self.boundary
        x_mid = (x_min + x_max) / 2
        y_mid = (y_min + y_max) / 2
        self.northwest = QuadTree((x_min, x_mid, y_min, y_mid), self.capacity)
        self.northeast = QuadTree((x_mid, x_max, y_min, y_mid), self.capacity)
        self.southwest = QuadTree((x_min, x_mid, y_mid, y_max), self.capacity)
        self.southeast = QuadTree((x_mid, x_max, y_mid, y_max), self.capacity)
        self.divided = True




    def insert(self, agent_id, position):
        if not self.contains(position):
            return False




        if len(self.agents) < self.capacity:
            self.agents.append((agent_id, position))
            return True
        else:
            if not self.divided:
                self.subdivide()




            if self.northwest.insert(agent_id, position):
                return True
            elif self.northeast.insert(agent_id, position):
                return True
            elif self.southwest.insert(agent_id, position):
                return True
            elif self.southeast.insert(agent_id, position):
                return True




    def contains(self, position):
        x, y = position
        x_min, x_max, y_min, y_max = self.boundary
        return x_min <= x <= x_max and y_min <= y <= y_max




    def query(self, boundary):
        result = []
        if not self.intersects(boundary):
            return result




        for agent_id, position in self.agents:
            if self.point_in_boundary(position, boundary):
                result.append((agent_id, position))




        if self.divided:
            result += self.northwest.query(boundary)
            result += self.northeast.query(boundary)
            result += self.southwest.query(boundary)
            result += self.southeast.query(boundary)




        return result




    def intersects(self, boundary):
        x_min, x_max, y_min, y_max = self.boundary
        bx_min, bx_max, by_min, by_max = boundary
        return not (bx_min > x_max or bx_max < x_min or by_min > y_max or by_max < y_min)




    def point_in_boundary(self, position, boundary):
        x, y = position
        bx_min, bx_max, by_min, by_max = boundary
        return bx_min <= x <= bx_max and by_min <= y <= by_max












# DQN Agent class
class DQNAgent:
    def __init__(self, state_size, action_size):
        self.state_size = state_size
        self.action_size = action_size
        self.gamma = 0.95  # Discount factor
        self.epsilon = 1.0  # Exploration rate
        self.epsilon_min = 0.1
        self.epsilon_decay = 0.995
        self.learning_rate = 0.001
        self.optimizer = tf.keras.optimizers.Adam(learning_rate=self.learning_rate)
        self.model = self._build_model()








    def _build_model(self):
      model = tf.keras.Sequential([
          tf.keras.layers.Input(shape=(self.state_size,)),
          tf.keras.layers.Dense(16, activation='relu'),  # Reduced neurons
          tf.keras.layers.Dense(self.action_size, activation='linear')
      ])
      model.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=self.learning_rate), loss='mse')
      return model










    def act(self, state):
        if np.random.rand() <= self.epsilon:
            return random.randrange(self.action_size)  # Explore
        state = np.array(state).reshape(1, -1).astype(np.float32)  # Ensure correct state shape and type
        q_values = self.model(state, training=False)
        return np.argmax(q_values[0].numpy())  # Exploit






    def train(self, state, action, reward, next_state, done):
      # Convert inputs to tensors
      state = np.array(state).reshape(1, -1).astype(np.float32)
      next_state = np.array(next_state).reshape(1, -1).astype(np.float32)


      # Calculate the target value (scalar)
      target = reward
      if not done:
          q_future = np.amax(self.model(next_state, training=False)[0].numpy())
          target = reward + self.gamma * q_future


      # Ensure the target is a tensor of shape (1,)
      target = tf.convert_to_tensor([[target]], dtype=tf.float32)  # Ensure 2D shape








      with tf.GradientTape() as tape:
          # Get the Q-values for the current state
          q_values = self.model(state, training=True)
          # Extract the Q-value for the specific action and ensure it has a matching shape
          q_value = tf.expand_dims(q_values[0, action], axis=0)  # Shape (1,) to match target


          # Calculate the loss between the target and the predicted Q-value
          loss = tf.keras.losses.MeanSquaredError()(target, q_value)


      # Update the model
      grads = tape.gradient(loss, self.model.trainable_variables)
      self.optimizer.apply_gradients(zip(grads, self.model.trainable_variables))








      # Decay epsilon
      if self.epsilon > self.epsilon_min:
          self.epsilon *= self.epsilon_decay




# CVRPTW Environment with QuadTree for collision detection
class CVRPTWEnvironment:
    def __init__(self, orders_df, depot):
        self.orders_df = orders_df
        self.depot = depot
        self.current_time = 0
        self.grid_size = 1
        self.quad_tree_capacity = 4
        self.agent_positions = {}
        self.boundary = (0, 100, 0, 100)  # Assuming a 100x100 grid
        self.quad_tree = None
        self.reset()

    def reset(self):
        self.current_time = 0
        self.agent_positions = {agent_id: (random.uniform(0, 100), random.uniform(0, 100))
                                for agent_id in range(len(self.orders_df))}
        self.quad_tree = QuadTree(self.boundary, self.quad_tree_capacity)
        for agent_id, position in self.agent_positions.items():
            self.quad_tree.insert(agent_id, position)
        self.total_reward = 0  # Reset total reward for each episode
        return self._get_state()

    def _get_state(self):
        states = []
        for agent_id in self.agent_positions:
            position = self.agent_positions[agent_id]
            distances = [np.linalg.norm(np.array(position) - np.array((row['x'], row['y'])))
                         for _, row in self.orders_df.iterrows()]
            states.append(np.array([position[0], position[1], self.current_time, min(distances)]))
        return np.array(states)

    # def step(self, actions):
    #     episode_reward = 0
    #     agents_reached_goal = 0
    #     collision_penalty = -1  # Further reduced penalty
    #     goal_reward = 10        # Reduced reward to keep within desired range
    #     scaling_factor = 0.5    # Adjust scaling factor as needed

    #     # Clear the QuadTree and reinsert agents
    #     self.quad_tree = QuadTree(self.boundary, self.quad_tree_capacity)
    #     for agent_id, action in enumerate(actions):
    #         current_position = self.agent_positions[agent_id]
    #         goal_position = (self.orders_df.iloc[agent_id]['x'], self.orders_df.iloc[agent_id]['y'])
    #         move_direction = np.array(goal_position) - np.array(current_position)
    #         move_direction = move_direction / np.linalg.norm(move_direction)
    #         new_position = (current_position[0] + move_direction[0] * self.grid_size,
    #                         current_position[1] + move_direction[1] * self.grid_size)

    #         # Insert new position into QuadTree for querying
    #         self.quad_tree.insert(agent_id, new_position)

    #         # Query nearby agents for potential collisions
    #         boundary = (new_position[0] - self.grid_size, new_position[0] + self.grid_size,
    #                     new_position[1] - self.grid_size, new_position[1] + self.grid_size)
    #         nearby_agents = self.quad_tree.query(boundary)

    #         collision = False
    #         for other_agent_id, other_position in nearby_agents:
    #             if agent_id != other_agent_id and np.linalg.norm(np.array(new_position) - np.array(other_position)) < self.grid_size:
    #                 episode_reward += collision_penalty
    #                 collision = True
    #                 break

    #         if not collision:
    #             self.agent_positions[agent_id] = new_position
    #             if np.linalg.norm(np.array(new_position) - np.array(goal_position)) < self.grid_size:
    #                 agents_reached_goal += 1
    #                 episode_reward += goal_reward

    #     # Apply scaling and cap total reward
    #     self.total_reward += episode_reward * scaling_factor
    #     if self.total_reward > 1000:
    #         self.total_reward = 1000  # Cap reward to maximum of 1000

    #     self.current_time += 1
    #     done = self.current_time >= 100  # Terminate after 100 steps
    #     success_rate = agents_reached_goal / len(self.agent_positions)

    #     # Ensure total reward is within 10-1000 range
    #     if self.total_reward < 10:
    #         self.total_reward = 10

    #     return self._get_state(), self.total_reward, done, success_rate
    def step(self, actions):
        episode_reward = 0
        agents_reached_goal = 0
        collision_penalty = -1  # Penalty for collision
        goal_reward = 10        # Reward for reaching goal
        scaling_factor = 0.05   # Smaller scaling factor for better range control

        # Clear the QuadTree and reinsert agents
        self.quad_tree = QuadTree(self.boundary, self.quad_tree_capacity)
        for agent_id, action in enumerate(actions):
            current_position = self.agent_positions[agent_id]
            goal_position = (self.orders_df.iloc[agent_id]['x'], self.orders_df.iloc[agent_id]['y'])
            move_direction = np.array(goal_position) - np.array(current_position)
            move_direction = move_direction / np.linalg.norm(move_direction)
            new_position = (current_position[0] + move_direction[0] * self.grid_size,
                            current_position[1] + move_direction[1] * self.grid_size)

            # Insert new position into QuadTree for querying
            self.quad_tree.insert(agent_id, new_position)

            # Query nearby agents for potential collisions
            boundary = (new_position[0] - self.grid_size, new_position[0] + self.grid_size,
                        new_position[1] - self.grid_size, new_position[1] + self.grid_size)
            nearby_agents = self.quad_tree.query(boundary)

            collision = False
            for other_agent_id, other_position in nearby_agents:
                if agent_id != other_agent_id and np.linalg.norm(np.array(new_position) - np.array(other_position)) < self.grid_size:
                    episode_reward += collision_penalty
                    collision = True
                    break

            if not collision:
                self.agent_positions[agent_id] = new_position
                if np.linalg.norm(np.array(new_position) - np.array(goal_position)) < self.grid_size:
                    agents_reached_goal += 1
                    episode_reward += goal_reward

        # Scale down the episode reward and add to the total reward
        scaled_episode_reward = episode_reward * scaling_factor
        self.total_reward += scaled_episode_reward

        # Divide total reward by 10 to bring it closer to the range [10, 1000]
        adjusted_reward = self.total_reward / 10
        if adjusted_reward > 1000:
            adjusted_reward = 1000
        elif adjusted_reward < 10:
            adjusted_reward = 10

        self.current_time += 1
        done = self.current_time >= 100  # Terminate after 100 steps
        success_rate = agents_reached_goal / len(self.agent_positions)

        return self._get_state(), adjusted_reward, done, success_rate

    # Federated learning training function (updated to average model weights)
    def federated_train(environment, agents, episodes=50, max_steps=100):
        for e in range(episodes):
            state = environment.reset()
            done = False
            step = 0
            total_rewards = []




            while not done and step < max_steps:
                actions = [agent.act(state[i]) for i, agent in enumerate(agents)]
                next_state, total_reward, done, success_rate = environment.step(actions)
                total_rewards.append(total_reward)




                # Update each agent with its reward and state transition
                for i, agent in enumerate(agents):
                    agent.train(state[i], actions[i], total_reward, next_state[i], done)




                state = next_state
                step += 1




            # Log the total reward per episode
            print(f'Episode {e + 1}/{episodes}, Success Rate: {success_rate * 100:.2f}%, Total Reward: {sum(total_rewards)}')




            # Federated learning: Average the weights of all agents at the end of each episode
            average_weights = [np.mean([agent.model.get_weights()[layer] for agent in agents], axis=0) for layer in range(len(agents[0].model.get_weights()))]
            for agent in agents:
                agent.model.set_weights(average_weights)








# Testing function with additional performance metrics
def test_trained_model(environment, agents, episodes=1, max_steps=100):
    total_rewards = []
    total_collisions = 0
    total_success_rate = 0
    episode_rewards = []
    episode_collisions = []


    for e in range(episodes):
        state = environment.reset()
        done = False
        episode_reward = 0
        episode_collision_count = 0
        all_paths = [[] for _ in range(len(agents))]


        for step in range(max_steps):
            actions = [agent.act(state[i]) for i, agent in enumerate(agents)]
            next_state, total_reward, done, success_rate = environment.step(actions)


            # Track total reward and collisions for the episode
            episode_reward += total_reward
            if total_reward < 0:  # Assuming negative rewards indicate collisions
                episode_collision_count += 1


            # Calculate success rate and store each agent's path
            for i, path in enumerate(all_paths):
                if success_rate > 0:
                    path.append((state[i][:2], step))  # Store position and timestamp (step)


            state = next_state
            if done:
                break


        # Collect episode-specific metrics
        episode_rewards.append(episode_reward)
        episode_collisions.append(episode_collision_count)


        total_collisions += episode_collision_count
        total_success_rate += success_rate


    # Plotting total rewards and collisions per episode
    plt.figure(figsize=(12, 5))


    # Total Rewards per Episode
    plt.subplot(1, 2, 1)
    plt.plot(range(episodes), episode_rewards, label="Total Reward per Episode")
    plt.xlabel("Episode")
    plt.ylabel("Total Reward")
    plt.title("Total Reward per Episode")
    plt.legend()


    # Collisions per Episode
    plt.subplot(1, 2, 2)
    plt.plot(range(episodes), episode_collisions, label="Collisions per Episode", color="red")
    plt.xlabel("Episode")
    plt.ylabel("Collisions")
    plt.title("Collisions per Episode")
    plt.legend()


    plt.tight_layout()
    plt.show()


    # Display final testing metrics
    print(f'Final Testing Results: Total Collisions: {total_collisions}, '
          f'Average Success Rate: {(total_success_rate / episodes) * 100:.2f}%')








# Visualization with timestamps
def visualize_paths(all_paths, environment, batch_size=10, agents=None):
    num_agents = len(all_paths)
    for batch_start in range(0, num_agents, batch_size):
        plt.figure(figsize=(10, 10))
        colors = cm.get_cmap('tab10', min(batch_size, num_agents - batch_start))
        for i in range(batch_size):
            agent_id = batch_start + i
            if agent_id >= num_agents:
                break
            path = all_paths[agent_id]




            # Plot the agent's path with timestamps
            path_positions = np.array(path)
            timestamps = np.arange(len(path_positions))




            plt.plot(path_positions[:, 0], path_positions[:, 1], marker='o', label=f'Agent {agent_id}', color=colors(i))




            # Plot the goal with a star marker in the same color as the agent
            goal_position = (environment.orders_df.iloc[agent_id]['x'], environment.orders_df.iloc[agent_id]['y'])
            plt.scatter(goal_position[0], goal_position[1], marker='*', s=200, color=colors(i), label=f'Goal {agent_id}')




            # # Annotate each step with the timestamp
            # for j in range(len(path_positions)):
            #     plt.annotate(f'Step {timestamps[j]}', (path_positions[j, 0], path_positions[j, 1]))




        plt.title(f'Agent Paths from {batch_start} to {min(batch_start + batch_size, num_agents)}')
        plt.xlabel('X Coordinate')
        plt.ylabel('Y Coordinate')
        plt.grid(True)
        plt.legend()
        plt.show()








# Main Execution
state_size = 4  # [x, y, time, min_distance]
action_size = 5  # 5 actions: up, down, left, right, stay




# Training
agents_train = [DQNAgent(state_size=state_size, action_size=action_size) for _ in range(len(train_df))]
env_train = CVRPTWEnvironment(train_df, depot=(0, 0))
federated_train(env_train, agents_train, episodes=50, max_steps=100)




# Testing
agents_test = [DQNAgent(state_size=state_size, action_size=action_size) for _ in range(len(test_df))]
env_test = CVRPTWEnvironment(test_df, depot=(0, 0))
test_trained_model(env_test, agents_test, episodes=50, max_steps=100)







```

```python
import os
import pandas as pd
import torch
import torch.nn as nn
import torch.optim as optim
import numpy as np
import matplotlib.pyplot as plt

# Load and preprocess the dataset
data_dir = '/content'  # Ensure this path is correct
orders_file = os.path.join(data_dir, 'orders.csv')
orders_df = pd.read_csv(orders_file)
orders_df = orders_df.head(10)  # Limit to 10 for now

# Normalize coordinates
def normalize_data(orders_df, grid_max=100):
    min_x, max_x = orders_df['x'].min(), orders_df['x'].max()
    min_y, max_y = orders_df['y'].min(), orders_df['y'].max()
    max_range = max(max_x - min_x, max_y - min_y)
    orders_df['x'] = ((orders_df['x'] - min_x) / max_range) * grid_max
    orders_df['y'] = ((orders_df['y'] - min_y) / max_range) * grid_max
    return orders_df

orders_df = normalize_data(orders_df)

# Define Actor and Critic networks
class Actor(nn.Module):
    def __init__(self, state_dim, action_dim):
        super(Actor, self).__init__()
        self.fc1 = nn.Linear(state_dim, 128)
        self.fc2 = nn.Linear(128, 128)
        self.fc3 = nn.Linear(128, action_dim)

    def forward(self, x):
        x = torch.relu(self.fc1(x))
        x = torch.relu(self.fc2(x))
        return torch.tanh(self.fc3(x))

class Critic(nn.Module):
    def __init__(self, state_dim, action_dim):
        super(Critic, self).__init__()
        self.fc1 = nn.Linear(state_dim + action_dim, 128)
        self.fc2 = nn.Linear(128, 128)
        self.fc3 = nn.Linear(128, 1)

    def forward(self, state, action):
        x = torch.cat([state, action], dim=1)
        x = torch.relu(self.fc1(x))
        x = torch.relu(self.fc2(x))
        return self.fc3(x)

# Define MADDPG
class MADDPG:
    def __init__(self, num_agents, state_dim, action_dim, lr=0.01, gamma=0.99, tau=0.01):
        self.actors = [Actor(state_dim, action_dim) for _ in range(num_agents)]
        self.critics = [Critic(state_dim * num_agents, action_dim * num_agents) for _ in range(num_agents)]
        self.actor_optimizers = [optim.Adam(actor.parameters(), lr=lr) for actor in self.actors]
        self.critic_optimizers = [optim.Adam(critic.parameters(), lr=lr) for critic in self.critics]
        self.gamma = gamma
        self.tau = tau
        self.num_agents = num_agents

    # Placeholder for train step
    def train(self, batch):
        pass  # Simplified: actual training logic would go here

# Initialize MADDPG
num_agents = len(orders_df)
state_dim = 2  # (x, y) coordinates
action_dim = 2  # (dx, dy) movement per step
maddpg = MADDPG(num_agents, state_dim, action_dim)

# Define target positions for each agent
target_positions = orders_df[['x', 'y']].values

# Training setup
episodes = 1000
rewards = []
collisions = []
# Training loop
for episode in range(episodes):
    total_reward = 0
    num_collisions = 0

    # Reset agent positions to random starting points
    agent_positions = np.random.rand(num_agents, state_dim) * 100

    for step in range(50):  # Number of steps per episode
        actions = []
        for i in range(num_agents):
            state = torch.FloatTensor(agent_positions[i])
            action = maddpg.actors[i](state).detach().numpy()
            actions.append(action)

        # Update agent positions based on actions
        agent_positions += np.array(actions)

        # Scale down the reward calculation
        step_reward = -0.01 * np.linalg.norm(agent_positions - target_positions, axis=1).sum()
        total_reward += step_reward

        # Check for collisions (distance < threshold)
        collisions_per_step = np.sum(np.linalg.norm(agent_positions[:, None] - agent_positions, axis=2) < 1) - num_agents
        num_collisions += max(0, collisions_per_step)

    rewards.append(total_reward)
    collisions.append(num_collisions)
    maddpg.train(None)


# Plotting results
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
plt.plot(range(episodes), rewards, label="MADDPG Reward per Episode")
plt.xlabel("Episode")
plt.ylabel("Reward")
plt.legend()

plt.subplot(1, 2, 2)
plt.plot(range(episodes), collisions, label="MADDPG Collisions per Episode")
plt.xlabel("Episode")
plt.ylabel("Collisions")
plt.legend()
plt.tight_layout()
plt.show()

```

```python
import os
import numpy as np
import pandas as pd
import random
import matplotlib.pyplot as plt
import torch
import torch.nn as nn
import torch.optim as optim


# Set random seed for reproducibility
np.random.seed(42)
torch.manual_seed(42)
random.seed(42)


# Load CVRPTW Dataset
data_dir = '/content'  # Ensure this path is correct
orders_file = os.path.join(data_dir, 'orders.csv')
orders_df = pd.read_csv(orders_file)
orders_df = orders_df.head(10)  # Limit to 10 for now


# Preprocess the dataset (normalize coordinates)
def normalize_data(orders_df, grid_max=100):
    min_x, max_x = orders_df['x'].min(), orders_df['x'].max()
    min_y, max_y = orders_df['y'].min(), orders_df['y'].max()
    max_range = max(max_x - min_x, max_y - min_y)
    orders_df['x'] = ((orders_df['x'] - min_x) / max_range) * grid_max
    orders_df['y'] = ((orders_df['y'] - min_y) / max_range) * grid_max
    return orders_df


orders_df = normalize_data(orders_df)


# Split the dataset into 75% training and 25% testing
train_size = int(0.75 * len(orders_df))
train_df = orders_df[:train_size]
test_df = orders_df[train_size:]


class ActorCritic(nn.Module):
    def __init__(self, state_size, action_size):
        super(ActorCritic, self).__init__()
        self.fc = nn.Sequential(
            nn.Linear(state_size, 128),
            nn.ReLU(),
            nn.Linear(128, 128),
            nn.ReLU()
        )
        self.actor = nn.Linear(128, action_size)
        self.critic = nn.Linear(128, 1)


    def forward(self, x):
        x = self.fc(x)
        return self.actor(x), self.critic(x)


class Agent:
    def __init__(self, state_size, action_size, lr=0.001):
        self.model = ActorCritic(state_size, action_size)
        self.optimizer = optim.Adam(self.model.parameters(), lr=lr)
        self.action_size = action_size


    def select_action(self, state):
        state_tensor = torch.FloatTensor(state)
        logits, _ = self.model(state_tensor)
        action_prob = torch.softmax(logits, dim=-1)
        action = torch.multinomial(action_prob, 1).item()
        return action


    def update(self, states, actions, rewards):
        states_tensor = torch.FloatTensor(states)
        actions_tensor = torch.LongTensor(actions)
        rewards_tensor = torch.FloatTensor(rewards)


        logits, state_values = self.model(states_tensor)
        action_log_probs = torch.log_softmax(logits, dim=-1)
        selected_log_probs = action_log_probs[range(len(actions)), actions_tensor]


        advantages = rewards_tensor - state_values.squeeze()


        actor_loss = -(selected_log_probs * advantages.detach()).mean()
        critic_loss = advantages.pow(2).mean()
        loss = actor_loss + critic_loss


        self.optimizer.zero_grad()
        loss.backward()
        self.optimizer.step()


class CVRPTWEnvironment:
    def __init__(self, data, grid_size=100):
        self.data = data
        self.grid_size = grid_size
        self.agent_positions = [(random.uniform(0, grid_size), random.uniform(0, grid_size)) for _ in range(len(data))]
        self.goals = list(zip(data['x'], data['y']))


    def reset(self):
        self.agent_positions = [(random.uniform(0, self.grid_size), random.uniform(0, self.grid_size)) for _ in range(len(self.data))]
        return [self._get_state(agent_id) for agent_id in range(len(self.agent_positions))]


    def _get_state(self, agent_id):
        agent_x, agent_y = self.agent_positions[agent_id]
        goal_x, goal_y = self.goals[agent_id]
        return np.array([agent_x, agent_y, goal_x, goal_y])

    def step(self, actions):
      rewards = []
      collisions = 0
      for agent_id, action in enumerate(actions):
          dx, dy = 0, 0
          if action == 0: dy = 1
          elif action == 1: dx = 1
          elif action == 2: dy = -1
          elif action == 3: dx = -1
          x, y = self.agent_positions[agent_id]
          new_x, new_y = x + dx, y + dy
          new_x, new_y = min(max(new_x, 0), self.grid_size), min(max(new_y, 0), self.grid_size)
          self.agent_positions[agent_id] = (new_x, new_y)
          goal_x, goal_y = self.goals[agent_id]
          distance = np.sqrt((new_x - goal_x) ** 2 + (new_y - goal_y) ** 2)

          # Base reward proportional to distance
          reward = -distance / self.grid_size
          # Bonus for reaching the goal
          if distance < 1.0: reward += 20
          # Penalty for collisions
          for other_id in range(len(self.agent_positions)):
              if other_id != agent_id:
                  ox, oy = self.agent_positions[other_id]
                  if np.sqrt((new_x - ox) ** 2 + (new_y - oy) ** 2) < 1.0:
                      reward -= 10
                      collisions += 1
          rewards.append(reward)
      return [self._get_state(agent_id) for agent_id in range(len(self.agent_positions))], rewards, collisions



def train_model(environment, agents, episodes=100):
    rewards_per_episode = []
    collisions_per_episode = []


    for episode in range(episodes):
        states = environment.reset()
        total_rewards = 0
        total_collisions = 0


        for _ in range(50):  # Limit each episode to 50 steps
            actions = [agent.select_action(state) for agent, state in zip(agents, states)]
            next_states, rewards, collisions = environment.step(actions)


            for agent, state, action, reward in zip(agents, states, actions, rewards):
                agent.update([state], [action], [reward])


            states = next_states
            total_rewards += sum(rewards)
            total_collisions += collisions


        rewards_per_episode.append(total_rewards)
        collisions_per_episode.append(total_collisions)
        print(f"Episode {episode+1}: Total Reward = {total_rewards}, Total Collisions = {total_collisions}")


    # Plotting the results
    plt.figure(figsize=(12, 6))
    plt.subplot(1, 2, 1)
    # plt.plot(range(episodes), rewards_per_episode, label='Reward per Episode')
    plt.plot(range(episodes), rewards_per_episode, label="Actor Critic Policy Reward per Episode", color='blue')
    plt.xlabel('Episodes')
    plt.ylabel('Total Reward')
    plt.title('Reward per Episode')
    plt.legend()


    plt.subplot(1, 2, 2)
    plt.plot(range(episodes), collisions_per_episode, label='Actor Critic Policy Collisions per Episode', color='red')

    plt.xlabel('Episodes')
    plt.ylabel('Total Collisions')
    plt.title('Collisions per Episode')
    plt.legend()


    plt.tight_layout()
    plt.show()


# Initialize environment and agents
environment = CVRPTWEnvironment(train_df)
agents = [Agent(state_size=4, action_size=4) for _ in range(len(train_df))]


# Train the model
train_model(environment, agents, episodes=100)



```

```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Load and preprocess the dataset
data_dir = '/content'  # Ensure this path is correct
orders_file = os.path.join(data_dir, 'orders.csv')
orders_df = pd.read_csv(orders_file)
orders_df = orders_df.head(10)  # Limit to 10 for now

# Normalize coordinates
def normalize_data(orders_df, grid_max=100):
    min_x, max_x = orders_df['x'].min(), orders_df['x'].max()
    min_y, max_y = orders_df['y'].min(), orders_df['y'].max()
    max_range = max(max_x - min_x, max_y - min_y)
    orders_df['x'] = ((orders_df['x'] - min_x) / max_range) * grid_max
    orders_df['y'] = ((orders_df['y'] - min_y) / max_range) * grid_max
    return orders_df

orders_df = normalize_data(orders_df)

# Define target positions for each agent
num_agents = len(orders_df)
state_dim = 2  # (x, y) coordinates
target_positions = orders_df[['x', 'y']].values

# Training setup
episodes = 100
rewards_random = []
collisions_random = []

# Random Policy Training loop
for episode in range(episodes):
    total_reward_random = 0
    num_collisions_random = 0
    # Reset agent positions to random starting points
    agent_positions = np.random.rand(num_agents, state_dim) * 100
    for step in range(50):  # Number of steps per episode
        # Random Policy actions
        actions_random = (np.random.rand(num_agents, state_dim) - 0.5) * 2  # Random values in range [-1, 1]
        # Update agent positions based on actions
        agent_positions += actions_random
        # Calculate rewards for Random Policy
        reward_random = -np.linalg.norm(agent_positions - target_positions, axis=1).sum() * 0.02  # Scale rewards
        total_reward_random += reward_random
        # Check for collisions (distance < threshold)
        collisions_per_step_random = np.sum(np.linalg.norm(agent_positions[:, None] - agent_positions, axis=2) < 1) - num_agents
        num_collisions_random += max(0, collisions_per_step_random)
    rewards_random.append(total_reward_random)
    collisions_random.append(num_collisions_random)

# Plotting results for Random Policy
plt.figure(figsize=(12, 5))
# Rewards
plt.subplot(1, 2, 1)
plt.plot(range(episodes), rewards_random, label="Random Policy Reward per Episode", color='blue')
plt.xlabel("Episode")
plt.ylabel("Reward")
plt.legend()
# Collisions
plt.subplot(1, 2, 2)
plt.plot(range(episodes), collisions_random, label="Random Policy Collisions per Episode", color='orange')
plt.xlabel("Episode")
plt.ylabel("Collisions")
plt.legend()
plt.tight_layout()
plt.show()
```
